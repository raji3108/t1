```python
!pip install transformers torch --quiet

from transformers import pipeline

# Load pre-trained summarization pipeline
summarizer = pipeline("summarization", model="facebook/bart-large-cnn")

# Example long article
input_text = """
Artificial Intelligence (AI) is a field of computer science that emphasizes the creation of intelligent machines
that work and react like humans. Some of the activities computers with artificial intelligence are designed for include:
Speech recognition, Learning, Planning, and Problem-solving. AI is continuously evolving to benefit many different industries.
Machines are wired using a cross-disciplinary approach based on mathematics, computer science, linguistics, psychology,
and more. Advances in AI technology are pushing the boundaries of how machines can be used in day-to-day life.
"""

# Generate and print summary
summary = summarizer(input_text, max_length=100, min_length=30, do_sample=False)
print("Original Text:\n", input_text)
print("\nSummary:\n", summary[0]['summary_text'])
```

```python

!pip install transformers torch --quiet

from transformers import pipeline

# Load pre-trained summarization pipeline
summarizer = pipeline("summarization", model="facebook/bart-large-cnn")

# Ask user for input
print("Enter a long paragraph to summarize (minimum 3-4 lines recommended):\n")
input_text = input()

# Check input length
if len(input_text.split()) < 50:
    print("\nText is too short. Please enter at least 50 words for effective summarization.")
else:
    # Generate and print summary
    summary = summarizer(input_text, max_length=100, min_length=30, do_sample=False)
    print("\n\nSummary:\n")
    print(summary[0]['summary_text'])
```
